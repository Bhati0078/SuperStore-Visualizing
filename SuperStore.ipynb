```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Read Data
df =pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Sample - Superstore (1).xls')
df
```

```python
# Check the column info
df.info()
```

```python
# Top 5 City where we have Maximum Sales
df.groupby(['State'])['Sales'].sum().sort_values(ascending=False)[0:5].plot(kind='bar')
```

```python
#sort the data by Order Date
df.sort_values('Order Date')
```

```python
# Make a New Column and extract the month and Year's
df['Month_Year'] = df['Order Date'].dt.strftime('%m-%Y')
df.head(1)
```

```python
# Sales by Months
a=df.groupby('Month_Year')['Sales'].sum()
pd.DataFrame(a).reset_index()
```

```python
# Monthly Sales
df.groupby('Month_Year')['Sales'].sum().plot()
```

```python
# Sales by Category and Sub-category
df.groupby(['Category','Sub-Category'])['Sales'].sum().plot(kind='bar')
```

```python
#
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df)
plt.show()
```

```python
# correlation between variables in a dataset
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

```python
# Segment by Counts
sns.countplot(x=df['Segment'])
```

```python
#relationships between variables in a dataset.
sns.pairplot(df)
```
